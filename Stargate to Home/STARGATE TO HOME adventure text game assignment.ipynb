```python
## ###########################STARGATE TO HOME#################################
############################################################################

##INTRODUCTION
#Constructing Introduction and explanation of the game 
def introduction():
    print(f"""STARGATE TO HOME
    
Hello and welcome to "Stargate to home" !!
    
This adventure text game is all about bringing your character to
the good old Planet Earth.
He / she after pushing himself to the edge of the Andromeda galaxy on a 
mission to explore deep space and new life forms, crashed his / her 
spaceship named "Valkyrion-13" (who the hell chose that number?!) on an
unknown planet.
Unfortunately, all means of communication are destroyed and the only way
to return is to reach what in the distance looks like a Stargate, a 
space-time passage, similar to that present in the Hult Boston military
base on Planet Earth.
A couple of indications before starting:
- every decision you make will have repercussions on your path;
- read the questions carefully;
    
That's it, let the adventure begin!
Good luck!\n""")

    input(prompt = "<Press enter to start the game>\n")
    phase_1() #This move us to the crash_site stage

##PHASE 1
#Constructing phase_1 stage, asking user name, describing the initial game scenario and first question
def phase_1():
    
    #Defining astronaut_name variable
    astronaut_name = input(prompt = "Before going on, what is your astronaut name?\n>")
    
    #Converting astronaut_name using .capitalize
    astronaut_name = astronaut_name.capitalize()
    
    print(f"""\nOh hello {astronaut_name}! It's a pleasure to meet you!\n""")
    
    #This will start Phase 1 
    input(prompt = "<Press enter to access PHASE 1>\n") 
   
    #Defining choice_1 variable
    choice_1 = input(prompt = '*'*85 + "\n\nPHASE 1: CRASH SITE\n\n" + f"""
You just woke up after the crash of your spaceship "Valkyrion-13".
While checking the spaceship status you have noticed that all the means of
communication are destroyed.
You have about 12h of oxygen left and, the last time you checked, the 
closest rescue pod was about two days away from your location. 
In the distance you see a figure that looks similar to the stargate located
in the Hult Boston military base.
    
    You only have two alternatives, what do you choose? (type 1 or 2):
    1. Repair the communication systems even though they are severely damaged
    2. Walk to the figure that looks like the Stargate
    > """)
    
    #Converting choice_1 using .casefold
    choice_1 = choice_1.casefold()
    
    #Creating nested conditional statement
    if choice_1 == '1' or 'one' in choice_1:
        print(f"""
Ok then, let's see what can be done......... Those tools definitely don't want 
to work and unfortunately you spent too much time trying to repair them. 
Now you have unsufficient oxygen to go to the stargate""")
        last_chance() #This move us to the last_chance stage
    
    elif choice_1 == '2' or 'two'in choice_1:
        print("""\nGood choice! You proceed directly to the Stargate position\n""")
        phase_2() #This will move us to the phase_2 stage
    
    else: 
        print("<Invalid entry please try again>")
        phase_1() #This will move us back to the phase_! first question

##LAST CHANCE
#Constructing last_chance stage to give a second chance to the user to proceed to the PHASE 2
def last_chance():
    
    #Defining last_choice variable
    last_choice = input(prompt= """
    What do you want to do? (please type 1 or 2)
    1.Search for another way to reach the Stargate
    2.keep trying with repairs
    >""")
    
    #Converting last_choice using .casefold()
    last_choice = last_choice.casefold()
    
    #Creating nested conditional statement
    if last_choice == '1' or 'one' in last_choice:
        print(f"""
Wait....In the starship's load was included a rover, let's see if
it is still working......\n""")
        
        #Defining rover_ignition variable
        rover_ignition = input(prompt="\n<Type 'start' to start the rover engine>\n")
        
        #Converting rover_recognition using .casefold
        rover_ignition = rover_ignition.casefold()
        
        #Creating nested conditional statement
        if rover_ignition == 'start':
            print(f"""
pressing engine start button <click>
................................................
trtrtrtrtrtrtrttrtr vrooooom vrooooooooooooom""")
            print("""\n
Awesome! It works! Now we can go to the Stargate!\n""")
            phase_2() #This will move us to the phase_2 stage
            
        else:
            print("\n<please answer the question again>\n")
            last_chance() #This will move us back to the last_chance first question
    
    elif last_choice == '2' or 'two' in last_choice:
        print(f"""\n
You're stubborn ah? but no way, the means of communication are
irreparable. I'm sorry but your oxygen supply is finished. You are dead""")
        fail() #This will move us to the fail stage
    
    else:
        print("\n<please try again and remember to type '1' or '2'>\n")
        last_chance() #This will move us back to the last_chance stage initial question

##FAIL END/RESTART GAME
#Constructing fail stage to restart or end the game
def fail():
    
    #Creating restart_exit variable
    restart_exit = input(prompt = """\nWANT TO RESTART THE GAME?\n<please type yes or no>\n>""")
    
    #Creating nested conditional statement
    if restart_exit == "yes" or 'Yes' in restart_exit:
        introduction() #This will move back to the introduction
    
    elif restart_exit == "no" or 'No'in restart_exit:
        print("Hope you enjoyed the game!") #This will end the game
    
    else:
        print("please type again")
        fail() #This will restart the fail function

##PHASE 2
#Constructing phase_2 stage
#In this stage we will ask the user to find a 3-digit code in 3 attempts using a while loop
def phase_2():
    
    input(prompt = """<Press enter to access PHASE 2>\n>""")
    
    print("*"*85 + """\n\nPHASE 2: STARGATE\n\n""" + f"""
Here we are, we have finally reached the Stargate. 
Unfortunately it seems it takes a combination of 3 numbers to open it, but we can get some 
help from our portable decipherer.\n""")
    
    input(prompt = "\n<press enter to activate it>\n>")
    
    print(f"""\n
Oh something went wrong, the device only partially decrypted the combination 
and we need to correct the errors. Here's the issue we have to solve to find the 
right 3-digits for the combination:

             682 --> one digit is right and in its place
             614 --> one digit is right but in the wrong place
             206 --> two digits are right but both are in the wrong place
             738 --> all digits are wrong
             380 --> one digit is right but in the wrong place""")

#Defining combination variable
    combination = '042'

#Defining combination_attempts
    combination_attempts = 3

#Creating while loop with nested conditionals
    while combination_attempts > 0:
        
        code_attempt = input(prompt = """
So what are the 3-digits of our code? You have 3 attempts to find the right combination.
(remember to type them in the right order)\n>""")
        if code_attempt == combination:
            print("""
            Congratulations the code is right! The Stargate starts working!""")
            phase_3() #This will move us to the next stage
            break #This will stop while loop from running
            
        elif code_attempt != combination and combination_attempts == 3:
            print("""
            The combination is wrong, I will give you a hint: the first digit is 0\n""")
            combination_attempts -= 1 #This will reduce the number of combination_attempts
            print(f"""You have {combination_attempts} attempt(s) remaining.""")
        
        elif code_attempt != combination and combination_attempts == 2:
            print("""
            The combination is wrong, I will give you another hint: the second digit is 4\n""")
            combination_attempts -= 1 #This will reduce the number of combination_attempts
            print(f"""You have {combination_attempts} attempt(s) remaining.""")
        
        elif code_attempt != combination and combination_attempts == 1:
            print("""
            Combination is wrong. The Stargate starts to overheat 
            .................. BOOOOOOOM ...................
            I'm sorry but you died in the explosion\n""")
            fail() #This will restart the fail function
        
        else:
            print(f"""
            Please type again and remember to use numbers, not words!""")

##PHASE 3 and ending the game
#Constructing phase_3 stage
#In this stage we will ask the user to enter the stargate, type the earth coordinates and finish the game
def phase_3():       
    
    input(prompt = """<Press enter to access PHASE 3>\n>""")
    
    print("*"*85 + """\n\nPHASE 3: BACK TO EARTH\n\n""" + f"""
Here we are, we have finally made the Stargate working. 
There is only one final step to take, which is to enter the coordinates of the 
Planet Earth in the 3 appropriate quadrants. But how can we go about finding them?
..............................................................................
..............................................................................
Oh right, we have the galactic gps tracking system built into the spacesuit.\n""")
    
    input(prompt = "\n<click enter to activate it>\n>")
    phase_3_1() #This will move to the next phase

#Constructing phase_3_1 to start the coordinates typing
def phase_3_1():
    print("""
Oh right, the system is working and it's showing us the coordinate, we just need 
to put them into the quadrants.""")

#Creating coordinates variable
    coordinates = input(prompt= f"""
Here they are:
Spatial Latitude = 73
Spatial Longitude = 12
Galactic Curvature = 567

(Please type only the numbers separated by commas like in this example: 23,42,436)\n>""")

#Creating nested conditionals  
    if coordinates == '73,12,567':
        print("""Done? perfect.
The Stargate start making noises..............................................
You get sucked into the portal................................................
Hey but this is the Hult Boston military base on Planet Earth! 
You did it! Finally back home!

Well done Astronaut!\n\n""" + '*'*30 +"YOU WIN" + '*'*30) #END OF THE GAME
        
    elif coordinates != '73,12,567':
        print("""
        Are you sure?
        1) Yes
        2) No""")
        
        #Creating wrong_coordinates variable
        wrong_coordinates = input(prompt= """\n(Please type the number or yes/no)
        >""")
        
        #Converting wrong_coordinates using .casefold
        wrong_coordinates = wrong_coordinates.casefold()
        
        #Creating nested conditionals
        if wrong_coordinates == '1' or 'yes' in wrong_coordinates:
            print("""
        Oh nooooooooooooooooooooooooooooo.............................
        ..............................................................
        The coordinates were wrong! The Stargate took us to Saturn!
        The oxygen supply is finished. You are dead.""")
            fail() #This will move us to fail stage
        
        elif wrong_coordinates != '2' or 'no' in wrong_coordinates:
            print("""Ok so let's type them again\n""")
            phase_3_1() #This will move us back to the phase_3_1 stage
        
        else:
            print("Please type again the coordinates")
            phase_3_1() #This will move us back to the phase_3_1 stage
        
    else:
        print("""
        <Please type again and remember to separate the coordinates with commas>""")
        phase_3_1() #This will move us back to the phase_3_1

        
#Defining the first function in our map that will start our game
introduction()


#END OF THE GAME CODE
```
