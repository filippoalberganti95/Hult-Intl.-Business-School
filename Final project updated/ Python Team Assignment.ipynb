Methodology    
The first step taken in this analysis was to select the country that best represents the combined average of the entire Western European section. This country was selected based on five categories that were deemed important representations of a country, these were derived from the UN sustainable development goals:
- Poverty -> According to the UN poverty can severely limit someone’s potential as well as reduce their quality of life.
- Education -> The UN categorizes education as a human right because it allows an individual to improve their life prospects and 
- Technological Infrastructure ->Technology is one of the major pillars for implementing a better world, due to its heavy association with innovation, business opportunities and development of trade. 
- Healthcare -> Important because it ensures a longer living population with a better standard of life 
- Economic - > A countries economic level, increases the possibility of an individual to a better standard of life, socially, economically and mentally.

Once the metrics for the five categories was defined and the best representative country was chosen, a comparison to Eastern Europe was conducted utilizing the same process as was conducted for Western Europe. This comparison was done in order to highlight the difference that is expected in different global regions.

# 3. Analysis Strategy:
- Step 1: ***Check the dataset***
- Step 2: ***Group the Columns by Category***
- Step 3: ***Filter the dataset***
- Step 4: ***Find the NULL Value***
- Step 5: ***Make a decision (Data Cleaning)***

     - Due to the presence of null values (missing inputs) and outliers this analysis took several steps in order to mitigate their impact on the final result, but it also sought to understand the reasoning for these outliers.
     - First missing values was checked to see if missing values represent over 5% of the data, in order to evaluate the reliability and accuracy of the data that is going to be processed. 
     - The missing values were then inputted with the median value of all the other countries. The median was used instead of mean, in order to have a consistent approach throughout all analyzed columns, but to also take into account plausible skewedness that might be present in the data.
     - Outliers were identified through the use of a Box Plot. Each outlier was individually analyzed and explained in order to understand the reason behind the data’s position in regards to the rest of the data set.


- Step 6: ***Analysis the data to take the average value in each category***
- Step 7: ***Pick top 10 Countries in each category***
- Step 8: ***Make the intersection among the categories to pick the country stand for Western Europe***
- Step 9: ***Find the outliers in each category and explain reason by research***
- Step 10: ***Make comparison to Eastern Europe***
- Step 11: ***Conclusion***

```python
def infrastructure():
    
    # Choose all columns in Infrastructure: Communications and Education Categories

    columns_list=["Country Name","Adjusted net enrollment rate, primary (% of primary school age children)",
                 "Internet users (per 100 people)","Literacy rate, adult total (% of people ages 15 and above)",
                 "Mobile cellular subscriptions (per 100 people)","Primary completion rate, total (% of relevant age group)",
                 "School enrollment, primary (% net)"]
    ds_we_f = ds_we.loc[: ,columns_list]

    # Generate the colomuns in Communication category
    c_list=["Country Name","Internet users (per 100 people)","Mobile cellular subscriptions (per 100 people)"]
    ds_we_f_c =ds_we_f.loc[:,c_list]

    # Check the Null Values
    ds_we_f_c.isnull().sum()

    # Check if missing values affect less than 5%
    print(ds_we_f_c.isnull().mean(axis=0).round(decimals =2))

    # Clean the data for communication coloumns(Drop the countries with missing values)
    ds_we_f_cc = pd.DataFrame.copy(ds_we_f_c)
    ds_we_c_dropped = ds_we_f_cc.dropna(subset=["Internet users (per 100 people)","Mobile cellular subscriptions (per 100 people)"])

    # Infrastructure: Communication for Internet 
    # Making distribution chart for Internet users (per 100 people) without missing values
    sns.distplot(a =ds_we_c_dropped["Internet users (per 100 people)"],
                bins ='fd',
                hist = True,
                kde = True,
                rug = False,
                color = "black")

    plt.title(label = "Distribution of Internet users (per 100 people)")

    # Add vertical lines for mean and median
    plt.axvline(x = ds_we_c_dropped["Internet users (per 100 people)"].mean(),
               color = "red")
    plt.axvline(x = ds_we_c_dropped["Internet users (per 100 people)"].median(),
               color = "blue")

    # Add legend
    plt.legend(labels = ["Mean","Median"])

    # These compile and display the plot so that it is formatted as expected
    plt.tight_layout()
    plt.show()

    #Imputing missing values with median and making distribution chart to compare and analysis
    # Median for Internet users
    internet_users_median =ds_we_f_c["Internet users (per 100 people)"].median()

    # Make a copy of dataset
    ds_we_internet_user_median = pd.DataFrame.copy(ds_we_f_c)

    # Filling color NAs with Median
    ds_we_internet_user_median["Internet users (per 100 people)"].fillna(value = internet_users_median,
                                                                     inplace = True)

    # Make the comparison 
    # Setting figure size
    fig, ax = plt.subplots(figsize = [6,4],sharex = True,sharey = True)

    # Histogram for Adjusted net enrollment rate, primary (% of primary school age children)
    sns.distplot(a =ds_we_c_dropped["Internet users (per 100 people)"],
                bins ='fd',
                hist = True,
                kde = True,
                rug = False,
                color = "black")

    # Titles and labels
    plt.title(label = "Distribution of internet users with imputting median value")
    plt.xlabel(xlabel = "Users" )
    plt.ylabel(ylabel = "Number")

    # Histogram for after fillna
    sns.distplot( a =ds_we_internet_user_median["Internet users (per 100 people)"],
                bins = 'fd',
                hist = True,
                kde = True,
                rug = False,
                color = "gray")

    # Titles and labels
    plt.title(label = "Distribution of internet users with imputting median value")
    plt.xlabel(xlabel = "Users" )
    plt.ylabel(ylabel = "Number")

    # Adding a legend
    plt.legend(labels =  ['original distribution',
                          'imputed distribution'])
    # Add vertical lines for mean and median
    plt.axvline(x = ds_we_internet_user_median["Internet users (per 100 people)"].mean(),
               color = "red")
    plt.axvline(x = ds_we_internet_user_median["Internet users (per 100 people)"].median(),
               color = "blue")

    # Add legend
    plt.legend(labels = ["Mean","Median"])
    # Compiling and displaying the plot
    plt.tight_layout()
    plt.show()

    #Analysis the New Dataset to get the top countries approach the average value
    # Add a column named "Mean of Internet Users"
    ds_we_internet_user_median["Mean of Internet Users"]=ds_we_internet_user_median["Internet users (per 100 people)"].mean()
    # Add one column named "absolute difference value for internet users"
    ds_we_internet_user_median["absolute difference value for internet users"] = ds_we_internet_user_median["Internet users (per 100 people)"]-ds_we_internet_user_median["Mean of Internet Users"]
    # Ranking by absolute value
    ds_we_internet_user_median.iloc[ds_we_internet_user_median['absolute difference value for internet users'].abs().argsort()]
    # Make a list of top countries 
    top_countries_internet_users = ["France","Liechtenstein","Austria","Monaco","Belgium","Faeroe Islands","Ireland","Germany","Greenland","United Kingdom"]

    print("--"*58)
    print(f"""***Conclusion***:
    The Top countries in Internet unser category are in the list of {top_countries_internet_users}""")
    print("--"*58)

    #Infrastructure: Communication for Mobile cellular subscriptions
    sns.distplot(a =ds_we_c_dropped["Mobile cellular subscriptions (per 100 people)"],
                bins ='fd',
                hist = True,
                kde = True,
                rug = False,
                color = "black")

    plt.title(label = "Distribution of Mobile cellular subscriptions (per 100 people) without missing values")

    # Add vertical lines for mean and median
    plt.axvline(x = ds_we_c_dropped["Mobile cellular subscriptions (per 100 people)"].mean(),
               color = "red")
    plt.axvline(x = ds_we_c_dropped["Mobile cellular subscriptions (per 100 people)"].median(),
               color = "blue")

    # Add legend
    plt.legend(labels = ["Mean","Median"])

    # These compile and display the plot so that it is formatted as expected
    plt.tight_layout()
    plt.show()

    #Imputing missing values with median and making distribution chart to compare and analysis
    # Mean for Mobile cellular subscriptions
    mobile_median =ds_we_f_c["Mobile cellular subscriptions (per 100 people)"].median()

    # Make a copy of dataset
    ds_we_mobile_median = pd.DataFrame.copy(ds_we_f_c)

    # Filling color NAs with Mean
    ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"].fillna(value = mobile_median,
                                                                     inplace = True)

    # Make the comparison 
    # Setting figure size
    fig, ax = plt.subplots(figsize = [6,4],sharex = True,sharey = True)

    # Histogram for Adjusted net enrollment rate, primary (% of primary school age children)
    sns.distplot(a =ds_we_c_dropped["Mobile cellular subscriptions (per 100 people)"],
                bins ='fd',
                hist = True,
                kde = True,
                rug = False,
                color = "black")

    # Titles and labels
    plt.title(label = "Distribution of Mobile cellular subscriptions with imputing median value")
    plt.xlabel(xlabel = "Users" )
    plt.ylabel(ylabel = "Number")

    # Histogram for after fillna
    sns.distplot( a =ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"],
                bins = 'fd',
                hist = True,
                kde = True,
                rug = False,
                color = "gray")

    # Titles and labels
    plt.title(label = "Distribution of Mobile cellular subscriptions with imputing median value")
    plt.xlabel(xlabel = "Users" )
    plt.ylabel(ylabel = "Number")

    # Adding a legend
    plt.legend(labels =  ['original distribution',
                          'imputed distribution'])
    # Add vertical lines for mean and median
    plt.axvline(x = ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"].mean(),
               color = "red")
    plt.axvline(x = ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"].median(),
               color = "blue")

    # Add legend

    # Compiling and displaying the plot
    plt.tight_layout()
    plt.show()

    # Analysis the New Dataset to get the top countries approach the average value
    # Add one column named mean of mobile subscriptions
    ds_we_mobile_median["mean of mobile subscriptions"] = ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"].mean()
    # Add a column named "difference value of mobile subscriptions"
    ds_we_mobile_median["difference value of mobile subscriptions"] = ds_we_mobile_median["mean of mobile subscriptions"]-ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"]
    # Ranking by absolute value
    ds_we_mobile_median.iloc[ds_we_mobile_median['difference value of mobile subscriptions'].abs().argsort()]
    # Make a list of top countries 
    top_countries_mobile = ["Faeroe Islands","Ireland","Switzerland","Iceland","Spain","Denmark","Belgium","United Kingdom","Netherlands","Greenland"]

    print("--"*58)
    print(f"""***Conclusion***:
    The Top countries in mobile subscriptions category are in the list of {top_countries_mobile}""")
    print("--"*58)

    # Make a boxplot for Infrastructure: Communication Category
    sns.boxplot(data= [ds_we_mobile_median["Mobile cellular subscriptions (per 100 people)"],ds_we_internet_user_median["Internet users (per 100 people)"]],
                orient = "h")

    plt.title(  label = "Boxplot for Infrastructure: Communication")
    plt.xlabel(xlabel = "Users")
    plt.show()
    print(f"""
    Outliers:
    The Outlers in Monaco in Mobile subsriptions because the sim card from France or Italy can be used in the Monaco. Because people travelling to other places. It's not necessary to buy the sim-card only used in Monaco. https://bit.ly/3n1wXew""")
infrastructure()
```
